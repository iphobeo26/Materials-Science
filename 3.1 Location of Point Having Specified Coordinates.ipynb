## Location of Point Having Specified Coordinates

<p> For the unit cell shown in the accompanying sketch $(a)$, locate the point having coordinates $\frac{1}{4}1\frac{1}{2}$</p>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import fractions
from itertools import product, combinations
```

### Solution

<p>From sketch above, edge legnths for this unit cell are as follows: $a$ = 0.48 nm, $b$ = 0.46 nm, and $c$ = 0.40 nm. Furhtermore, in light of the preceding discussioin, the three point coordinate indices are $q$ = $\frac{1}{4}$, $r$ = 1, and $s$ = $\frac{1}{2}$. We use Equations 3.1a through 3.1c to determine lattice positions for this point as follows: </p>
<ul>
    <li>lattice positions referenced to the $x$ axis = $qa$ = $(\frac{1}{4})a$ = $\frac{1}{4}$(0.48 nm) = 0.12 nm
    <li>lattice positions referenced to the $y$ axis = $rb$ = $(1)b$ = (1)(0.46 nm) = 0.46 nm
    <li>lattice positions referenced to the $z$ axis = $sc$ = $(\frac{1}{2})c$ = $\frac{1}{2}$(0.40 nm) = 0.20 nm
</ul>
<p>To locate the point having these coordinatets within the unit cell, first use the $x$ lattice position and move from the origin (point $M$) 0.12 nm units along the $x$ axis (to point $N$), as shown in ($b$). Similarly, using the $y$ lattice position, proceed 0.46 nm parallel to the $y$ axis, from point $N$ to point $O$. Finally, move from this position 0.20 nm units parallel to the $z$ axis to point $P$ (per the $z$ lattice position), as noted in($b$). Thus, point $P$ corresponds to the $\frac{1}{4}1\frac{1}{2}$ point coordinates.</p>
        

```python
# Basic Parameters 
# -----------------------------------------------------
q = float(input("Insert the point coordinates at q: "))    # Point coordinates at q (float or int type)
r = float(input("Insert the point coordinates at r: "))    # Point coordinates at r (float or int type)
s = float(input("Insert the point coordinates at s: "))   # Point coordinates at s (float or int type)
a = float(input("Insert the length of x axis: "))          # Length of x axis
b = float(input("Insert the length of y axis: "))          # Length of y axis
c = float(input("Insert the length of z axis: "))          # Length of z axis

# Determine the lattice position
# -----------------------------------------------------
x_axis = q * a
y_axis = r * b
z_axis = s * c

# Printing the lattice position
# -----------------------------------------------------
print("Lattice position to the x axis: ", x_axis)
print("Lattice position to the y axis: ", y_axis)
print("Lattice position to the z axis: ", z_axis)
```
